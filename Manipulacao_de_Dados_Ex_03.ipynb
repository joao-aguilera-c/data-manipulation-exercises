#### Tarefa 1. Importe o dataset e salve os dados em um dataframe

Os dados estão salvos no arquivo ***datasets/users_dataset.txt***.\
Salve os dados em uma variável de nome *users*.

````python
import pandas as pd
users = pd.read_csv('datasets/users_dataset.csv', sep='|')
users
````

#### Tarefa 2. Descubra qual a idade média por ocupação.

*Dica: use a função groupby do Pandas na coluna occupation*

````python
idade_media_por_ocupacao = users.groupby('occupation').mean()['age']
idade_media_por_ocupacao
````

#### Tarefa 3. Para cada ocupação descubra a idade mínima e máxima.

````python
idade_minima_por_ocupacao = users.groupby('occupation').min()['age']
idade_maxima_por_ocupacao = users.groupby('occupation').max()['age']
# merge dos resultados
resultado = pd.merge(idade_minima_por_ocupacao, idade_maxima_por_ocupacao, left_index=True, right_index=True)
resultado.columns = ['idade_minima', 'idade_maxima']
resultado
````

#### Tarefa 5. Para cada ocupação calcule a porcentagem de homens e mulheres.

Exemplo:

```
occupation     gender
administrator  F          45.569620
               M          54.430380
artist         F          46.428571
               M          53.571429
doctor         M         100.000000
educator       F          27.368421
               M          72.631579
```

````python
# percentage of each gender by occupation:
gender_percentage_per_occupation = users.groupby(['occupation', 'gender']).count()
gender_percentage_per_occupation['percentage'] = None

gender_percentage_per_occupation

    
````

````python
i = 0

for index, row in gender_percentage_per_occupation.iterrows():
    if i % 2 == 0:
        gender_f = row.name[1]
        occupation = row.name[0]
        qty_f = row.user_id
    else:
        gender_m = row.name[1]
        qty_m = row.user_id
        # print(gender_m, qty_m, gender_f, qty_f)

        gender_percentage_per_occupation['percentage'].loc[
            gender_percentage_per_occupation.index == (occupation, gender_f)
        ] = (qty_f / (qty_f+qty_m)) *100
        gender_percentage_per_occupation['percentage'].loc[
            gender_percentage_per_occupation.index == (occupation, gender_m)
        ] = (qty_m / (qty_f+qty_m)) *100
    i +=1

gender_percentage_per_occupation['percentage']
````
